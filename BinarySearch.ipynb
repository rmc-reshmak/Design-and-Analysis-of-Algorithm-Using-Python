Write a Python program to accept n number of elements from user and element to be searched. Using Binary Search find element in an array.

```python
def binary_search(arr, key):
    low = 0
    high = len(arr) - 1

    while low <= high:
        mid = (low + high) // 2

        if arr[mid] == key:
            return mid
        elif arr[mid] < key:
            low = mid + 1
        else:
            high = mid - 1

    return -1

n = int(input("Enter the number of elements: "))
arr = []
print("Enter the array elements in sorted order:")
for i in range(n):
    element = int(input())
    arr.append(element)

key = int(input("Enter the element to search: "))
result = binary_search(arr, key)
if result != -1:
    print(key, "found at index", result)
else:
    print(key, "not found in the array")
```
