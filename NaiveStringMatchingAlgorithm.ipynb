Write a Python program to accept string and pattern to be serached in a string from user and find all occurance of pattern in string using Naive String Matching Algorithm.

```python
def naive_string_match(text, pattern):
    n = len(text)
    m = len(pattern)
    positions = []
    for i in range(n - m + 1):
        if text[i:i + m] == pattern:
            positions.append(i)

    return positions        

text = input("Enter the text: ")
pattern = input("Enter the pattern: ")

result = naive_string_match(text, pattern)

if len(result) > 0:
    print("Pattern found at indices:", result)
else:
    print("Pattern not found")
```
