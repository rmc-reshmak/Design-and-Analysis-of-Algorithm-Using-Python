Write a Python program to accept string and pattern to be searched in a string from user and find all occurance of pattern in string using Rabin-Karp Algorithm. (Character-By-Character Implementation)

<h1>Rabin-Karp Algorithm</h1>

The <b>Rabin-Karp Algorithm</b> is a string matching algorithm that uses hash values to search for a pattern in a text.

Instead of comparing the pattern with every substring character by character (like the Naive algorithm), it first compares their hash values.

<ul>
    <li>If the hash values are different, the substring cannot match the pattern.</li>
    <li>If the hash values are the same, it performs a character-by-character comparison to confirm the match (because different strings can sometimes have the same hash value, known as a hash collision).</li>
</ul>

<title>Rabin-Karp Algorithm</title>
<h2>Steps of the Rabin-Karp Algorithm</h2>
    <ol>
        <li>Calculate the hash value of the pattern.</li>
        <li>Calculate the hash value of the first substring of the text having the same length as the pattern.</li>
        <li>Compare the hash value of the pattern with the hash value of the current text window.</li>
        <li>If both hash values are equal, compare the characters of the pattern and the text one by one.</li>
        <li>If all characters match, the pattern is found.</li>
        <li>Move the search window one character to the right.</li>
        <li>Update the hash value of the new text window using the rolling hash technique instead of calculating it from scratch.</li>
        <li>Repeat the above steps until the end of the text.</li>
    </ol>

```python
def rabin_karp(text, pattern):
    d = 256      # Number of ASCII characters
    q = 101      # Prime number

    n = len(text)
    m = len(pattern)

    p = 0        # Hash value of pattern
    t = 0        # Hash value of text window
    h = 1

    # Calculate h = (d^(m-1)) % q
    for i in range(m - 1):
        h = (h * d) % q

    # Calculate initial hash values of pattern and first text window
    for i in range(m):
        p = (d * p + ord(pattern[i])) % q
        t = (d * t + ord(text[i])) % q

    # Slide the pattern over the text
    for i in range(n - m + 1):

        # If hash values match
        if p == t:

            # Compare characters one by one
            j = 0
            while j < m:
                if text[i + j] != pattern[j]:
                    break
                j += 1

            # If all characters matched
            if j == m:
                print("Pattern found at index", i)

        # Calculate hash value for next window
        if i < n - m:
            t = (d * (t - ord(text[i]) * h) + ord(text[i + m])) % q

            # Convert negative hash value to positive
            if t < 0:
                t += q

text = input("Enter the text: ")
pattern = input("Enter the pattern: ")

if len(pattern) > len(text):
    print("Pattern length should not be greater than text length.")
else:
    rabin_karp(text, pattern)
```
