Write a Python program to accept n number of elements from user and element to be searched. Using Linear Search find element in an array

```python
def linear_search(arr, key):
    for i in range(len(arr)):
        if arr[i] == key:
            return i
    return -1
    
n = int(input("Enter the number of elements: "))
arr = []
print("Enter the array elements:")
for i in range(n):
    element = int(input())
    arr.append(element)

key = int(input("Enter the element to search: "))
result = linear_search(arr, key)
if result != -1:
    print(key, "found at index", result)
else:
    print(key, "not found in the array")
```
